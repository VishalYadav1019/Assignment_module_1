```python
a=np.array([21,12,11,23,16,8])
b=np.array([4,23,4,6,72,6])
plt.plot(a,b, marker="*")  # marker function means mark the point with star...
```

```python
plt.plot(a,b,"*:g")  #give the colour g means green..
```

```python
plt.plot(a,b,"*:g" , ms = 12) # MS means marker size which is 12...
```

```python
plt.plot(a,b,"*:g" , ms = 20, mec= "r")  #mec means marker border colour which is given by red short
```

```python
#  mfc(marker face colour) which means inside colour if the marker..
plt.plot(a,b,"*:g" , ms = 20, mec= "r" ,mfc="pink")
```

```python
plt.plot(a,b, linestyle = "--")
```

```python
# giving the linewidth as 10...
plt.plot(a,b,linestyle = "--", c="black",linewidth = "10")
```

```python
#plot a and b 
plt.plot(a,linestyle = "-." , c = "g" , linewidth = "5.26")
plt.plot(b,linestyle = "-." , c = "r" , linewidth = "5.26")

plt.grid()
plt.xlabel("x") # gave the x axis label
plt.ylabel("y")  #gave the x axis label
plt.title("Graph")  # giving the title
plt.legend(a)   # legend means describing the element in the graph..
plt.legend(b)
```

```python
plt.bar(a,b, color="green", width = 0.4) 
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Bar Graph")
plt.legend(a)
```
